```python
# Filter data for meditation sessions, explicitly creating a copy
meditation_data = data[data['Title'].isin(meditation_titles)].copy()
```

```python
meditation_data['Finished On'] = pd.to_datetime(meditation_data['Finished On'])
```

```python
meditation_data.head()
```

```python
meditation_2023 = meditation_data[meditation_data['Finished On'].dt.year == 2023]
```

```python
meditation_daily = meditation_2023.groupby('Finished On').sum(numeric_only=True)
```

```python
plt.figure(figsize=(15, 6))
plt.plot(meditation_daily.index, meditation_daily['Duration']/60, label='Meditation Duration (minutes)')
plt.title('Daily Meditation Time in 2023')
plt.xlabel('Date')
plt.ylabel('Duration (minutes)')
plt.legend()
plt.grid(True)
plt.show()
```
